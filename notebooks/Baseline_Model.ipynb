```python
df = pd.read_csv("drive/MyDrive/feature_set_new.csv").drop(columns=["Unnamed: 0"])
df = df.dropna()
df = df.sample(frac=1)
x = df.drop('label', axis=1)
x = x.drop("orig_str", axis=1)
y = df['label']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

# Fit Random Forest classifier
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(x_train, y_train)

# Fit Decision Tree classifier
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(x_train, y_train)

# Fit KVM classifier
kvm = SVC(random_state=42)
kvm.fit(x_train, y_train)

# Fit XGBoost classifier
xgboost = XGBClassifier(random_state=42)
xgboost.fit(x_train, y_train)
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predict using each model
y_pred_rf = random_forest.predict(x_test)
y_pred_dt = decision_tree.predict(x_test)
y_pred_kvm = kvm.predict(x_test)
y_pred_xgb = xgboost.predict(x_test)

# Calculate metrics for each model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)

accuracy_kvm = accuracy_score(y_test, y_pred_kvm)
precision_kvm = precision_score(y_test, y_pred_kvm)
recall_kvm = recall_score(y_test, y_pred_kvm)
f1_kvm = f1_score(y_test, y_pred_kvm)

accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
precision_xgb = precision_score(y_test, y_pred_xgb)
recall_xgb = recall_score(y_test, y_pred_xgb)
f1_xgb = f1_score(y_test, y_pred_xgb)

Print the metrics for each model
print("Random Forest:")
print("- Accuracy:", accuracy_rf)
print("- Precision:", precision_rf)
print("- Recall:", recall_rf)
print("- F1 Score:", f1_rf)

print("\nDecision Tree:")
print("- Accuracy:", accuracy_dt)
print("- Precision:", precision_dt)
print("- Recall:", recall_dt)
print("- F1 Score:", f1_dt)

print("\nKVM:")
print("- Accuracy:", accuracy_kvm)
print("- Precision:", precision_kvm)
print("- Recall:", recall_kvm)
print("- F1 Score:", f1_kvm)

print("\nXGBoost:")
print("- Accuracy:", accuracy_xgb)
print("- Precision:", precision_xgb)
print("- Recall:", recall_xgb)
print("- F1 Score:", f1_xgb)
```
